```python
plt.title("2020년 게임 시장 매출액")
labels = ['PC','모바일','콘솔','기타']
data = [49.012,108.311,10.925,20.597]
colors=['y','r','b','g']
explode = [0.1,0,0,0]
plt.pie(data, 
        explode=explode,
        labels=labels,
        colors=colors,
        autopct="%1.1f%%",
        startangle=30
        )
plt.show()
```
